# 1.WHICH BRANCH MAKES THE MOST SALES

```python
supmar.groupby('Branch')['Sales'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
branch_sales=supmar.groupby('Branch')['Sales'].sum().reset_index()
branch_sales

branch_sales.plot(kind='bar',x='Branch',y='Sales')
plt.show()
```

# 7.WHAT IS THE AVERAGE RATING OF PRODUCT LINE

```python
prorat=supmar.groupby('Product line')['Rating'].mean()
prorat
```

```python
prorat.reset_index()
prorat.plot(kind='line',x='Product lines',y='Rating',color='g')
plt.xticks(rotation=30)
plt.title('Average rating of product lines')
plt.ylabel('rating')
plt.legend()
plt.show()
```

# 8.WHAT ARE THE SALES TREND OVER TIME

```python
supmar.info()
```

```python
supmar['Date']=pd.to_datetime(supmar['Date'])
supmar.info()
```

```python
saltre=supmar.groupby('Date')['Sales'].sum()
```

```python
saltre.reset_index()
```

# 13.WHICH PRODUCT LINE HAS THE HIGHEST AVERAGE GROSS INCOME PER SALE?

# 14.IS THERE ANY CORRELATION BETWEEN CUSTOMER RATING AND GROSS INCOME?

```python
cor=supmar[['Rating','gross income']].corr()
cor
```

```python
plt.scatter(supmar['Rating'], supmar['gross income'], alpha=0.5, color='purple')
plt.title('Customer Rating vs Gross Income')
plt.xlabel('Rating')
plt.ylabel('Gross Income')
plt.tight_layout()
plt.show()
```
